```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
```

```python
# define Guassian Naive Model
model=DecisionTreeClassifier()
# perform k-fold cross validation with k=8
scores=cross_val_score(model,iris.data,iris.target,cv=8,scoring='accuracy')
# print the scores for each fold and the mean score
print("Scores for each fold: ",scores)
print("Mean score: ",scores.mean())
print("Standard deviation: ",scores.std())
```

```python
# plot the results
plt.figure(figsize=(8, 6))
plt.plot(range(1,9),scores,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.title('Cross Validation Scores')
plt.xlabel('Number of Folds')
plt.ylabel('Accuracy')
plt.ylim([0.5,1.0])
plt.show()
```
