# Dealing with Missing **Values**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset('titanic')
data.head()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(data.isnull(),cbar=False)
plt.show()
```

```python
data.info()
```

```python
print("----------------------------------------------------")
print(f"Missing values in each column:\n{data.isnull().sum().sort_values(ascending=False)}")
print("----------------------------------------------------")
print(f"Percentage of missing values in each column:\n{round(data.isnull().sum()/len(data)*100,2).sort_values(ascending=False)}")
```

```python
print(round(data.age.mean(),2))
```

```python
print(round(data.age.median(),2))
```

```python
# mode are often use for categorical data
print(round(data.age.mode(),2))
```

```python
data.age=data.age.fillna(data.age.median())
```

```python
print("----------------------------------------------------")
print(f"Missing values in each column:\n{data.isnull().sum().sort_values(ascending=False)}")
print("----------------------------------------------------")
print(f"Percentage of missing values in each column:\n{round(data.isnull().sum()/len(data)*100,2).sort_values(ascending=False)}")
```

```python
# drop column if not important
data.drop('deck',axis=1,inplace=True)
```

```python
print("----------------------------------------------------")
print(f"Missing values in each column:\n{data.isnull().sum().sort_values(ascending=False)}")
print("----------------------------------------------------")
print(f"Percentage of missing values in each column:\n{round(data.isnull().sum()/len(data)*100,2).sort_values(ascending=False)}")
```

```python
data.embark_town.value_counts()
```

```python
data.embark_town.mode()
```

```python
# replace with mode because it is non numerical or categorical
data.embark_town=data.embark_town.fillna(data.embark_town.mode()[0])
```

```python
print("----------------------------------------------------")
print(f"Missing values in each column:\n{data.isnull().sum().sort_values(ascending=False)}")
print("----------------------------------------------------")
print(f"Percentage of missing values in each column:\n{round(data.isnull().sum()/len(data)*100,2).sort_values(ascending=False)}")
```

```python
# replace with mode because it is non numerical or categorical
data.embarked=data.embarked.fillna(data.embarked.mode()[0])
```

```python
print("----------------------------------------------------")
print(f"Missing values in each column:\n{data.isnull().sum().sort_values(ascending=False)}")
print("----------------------------------------------------")
print(f"Percentage of missing values in each column:\n{round(data.isnull().sum()/len(data)*100,2).sort_values(ascending=False)}")
```

```python
# forward fill and backward fill
# data.age=data.age.ffill()
data.age=data.age.bfill()
```

```python
print("----------------------------------------------------")
print(f"Missing values in each column:\n{data.isnull().sum().sort_values(ascending=False)}")
print("----------------------------------------------------")
print(f"Percentage of missing values in each column:\n{round(data.isnull().sum()/len(data)*100,2).sort_values(ascending=False)}")
```

```python
# using KNN imputer
from sklearn.impute import KNNImputer
imputer=KNNImputer(n_neighbors=5)
```

```python
data.age=imputer.fit_transform(data[['age']])
```

```python
print("----------------------------------------------------")
print(f"Missing values in each column:\n{data.isnull().sum().sort_values(ascending=False)}")
print("----------------------------------------------------")
print(f"Percentage of missing values in each column:\n{round(data.isnull().sum()/len(data)*100,2).sort_values(ascending=False)}")
```

```python
# delete missing values rows
data=data.dropna()
```

```python
print("----------------------------------------------------")
print(f"Missing values in each column:\n{data.isnull().sum().sort_values(ascending=False)}")
print("----------------------------------------------------")
print(f"Percentage of missing values in each column:\n{round(data.isnull().sum()/len(data)*100,2).sort_values(ascending=False)}")
```
