# Data **Merging**

```python
import pandas as pd
```

```python
data1=pd.DataFrame({
    'id':[1,2,3],
    'name':['Ahmed','Fatima','Ali'],
    'age':[28,32,24]
})
data2=pd.DataFrame({
    'id':[1,2,3],
    'city':['Karachi','Lahore','Islamabad'],
    'occupation':['Engineer','Doctor','Teacher']
})
data1
```

```python
data2
```

```python
merged_data= pd.merge(data1,data2,on='id',how='inner')
merged_data
```

# Data **Concatenation**

```python
df1=pd.DataFrame({'id':[1,2,3],'name':['Ahmed','Fatima','Ali'],'age':[28,32,24]})
df2=pd.DataFrame({'id':[1,2,3],'name':['Sara','Usman','Aisha'],'age':[29,35,27]})
df3=pd.DataFrame({'id':[1,2,3],'city':['Karachi','Lahore','Islamabad'],'Province':['Punjab','Sindh','Islamabad']})
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# Concatenating Rows
concat_rows=pd.concat([df1,df2])
concat_rows
```

```python
# Concatenating Columns
concat_col=pd.concat([df1,df3],axis=1)
concat_col
```
