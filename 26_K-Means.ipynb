```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
```

```python
df=sns.load_dataset('diamonds')
df.head()
```

```python
plt.scatter(df['carat'],df['price'])
plt.xlabel('carat')
plt.ylabel('price')
plt.show()
```

```python
x=df[['carat','price']]
model=KMeans(n_clusters=4)
model.fit(x)
# Get the predicted labels clusters
labels=model.labels_
# Add labels in new column
df['labels']=labels
df.head()
```

```python
df.labels.unique()
```

```python
# plot the data with different colors for each cluster
plt.scatter(df['carat'],df['price'],c=df['labels'])
plt.xlabel('carat')
plt.ylabel('price')
plt.title('K-Means Clustering on Diamonds Dataset')
plt.show()
```

# **Practice on Primary data**

```python
df=pd.DataFrame({
    'Height':[185,170,168,179,182,188,180,180,180,180,177],
    'Weight':[72,56,60,88,72,77,71,70,84,88,67]
})
df
```

```python
plt.scatter(df['Height'],df['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
x=df[['Height','Weight']]
model=KMeans(n_clusters=2)
model.fit(x)
# Get the predicted labels clusters
labels=model.labels_
# Add labels in new column
df['labels']=labels
df.head()
```

```python
df.labels.unique()
```

```python
# plot the data with different colors for each cluster
plt.scatter(df['Height'],df['Weight'],c=df['labels'])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('K-Means Clustering on Primary Dataset')
plt.show()
```
